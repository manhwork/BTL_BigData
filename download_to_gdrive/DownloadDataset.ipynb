# Tải raw ReviewData

```python
from huggingface_hub import hf_hub_download
import os

repo_id = "McAuley-Lab/Amazon-Reviews-2023"

base = "/content/drive/MyDrive/amazon_reviews_2023"

categories = [
"Home_and_Kitchen",
"Tools_and_Home_Improvement",
"Patio_Lawn_and_Garden",
"Pet_Supplies",
"Appliances",
"Arts_Crafts_and_Sewing"
]

for cat in categories:

    print(f"\n===== DOWNLOADING {cat} =====")

    os.makedirs(f"{base}/{cat}", exist_ok=True)

    hf_hub_download(
        repo_id=repo_id,
        filename=f"raw/review_categories/{cat}.jsonl",
        repo_type="dataset",
        local_dir=f"{base}/{cat}",
        resume_download=True
    )

print("\n===== ALL DOWNLOADS FINISHED =====")
```

# Tải raw MetaData

```python
from huggingface_hub import hf_hub_download
import os

repo_id = "McAuley-Lab/Amazon-Reviews-2023"

categories = [
"Home_and_Kitchen",
"Tools_and_Home_Improvement",
"Patio_Lawn_and_Garden",
"Pet_Supplies",
"Appliances",
"Arts_Crafts_and_Sewing"
]

base = "/content/drive/MyDrive/amazon_reviews_2023"

for cat in categories:

    print("Downloading metadata:", cat)

    hf_hub_download(
        repo_id=repo_id,
        filename=f"raw/meta_categories/meta_{cat}.jsonl",
        repo_type="dataset",
        local_dir=f"{base}/{cat}",
        resume_download=True
    )

print("Metadata download finished")
```
